```python
# 1. Import required library
import pandas as pd
```

```python
# 2. Load the Titanic dataset
# Keep this CSV file in the same folder as this notebook.
df = pd.read_csv("titanic.csv(2).csv")

# Display first 5 rows
df.head()
```

```python
# 3. Check number of rows and columns
df.shape
```

```python
# 4. Check basic information about dataset
df.info()
```

```python
# 5. Check missing values in each column
df.isnull().sum()
```

```python
# 6. Check duplicate rows
df.duplicated().sum()
```

```python
# 7. Remove duplicate rows
df = df.drop_duplicates()

# Check shape after removing duplicates
df.shape
```

```python
# 8. Fill missing values

# Fill missing Age values with median age
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill missing Embarked values with most frequent value
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
# 9. Drop Cabin column
# Cabin has many missing values, so we remove it.
df = df.drop(columns=["Cabin"])
```

```python
# 10. Rename columns for better readability
df = df.rename(columns={
    "PassengerId": "Passenger_ID",
    "Survived": "Survival_Status",
    "Pclass": "Passenger_Class",
    "SibSp": "Siblings_Spouses",
    "Parch": "Parents_Children"
})

df.head()
```

```python
# 11. Convert data types

df["Sex"] = df["Sex"].astype("category")
df["Embarked"] = df["Embarked"].astype("category")
df["Passenger_Class"] = df["Passenger_Class"].astype("category")
df["Survival_Status"] = df["Survival_Status"].astype("int64")

df.info()
```

```python
# 12. Check missing values again after cleaning
df.isnull().sum()
```

```python
# 13. Final cleaned dataset preview
df.head()
```

```python
# 14. Save cleaned dataset
df.to_csv("cleaned_titanic.csv", index=False)

print("Cleaned dataset saved successfully as cleaned_titanic.csv")
```

## Cleaning Summary

- Missing values in **Age** were filled using the median age.
- Missing values in **Embarked** were filled using the most frequent value.
- **Cabin** column was removed because it had many missing values.
- Duplicate rows were checked and removed.
- Important columns were renamed for better readability.
- Categorical columns were converted to category data type.
- Final cleaned dataset was saved as **cleaned_titanic.csv**.

## Result

The Titanic dataset is now clean and ready for analysis.
